#### Funciones para convertir txt a csv y procesar columnas

```python
def replace_characters_dcgan(path):
    df = pd.read_csv(path, encoding="utf-8", sep=" ", header=None)
    df = df.drop(columns=[0, 1, 2, 4, 6, 7, 8, 9, 10, 11])
    df = df.rename(columns={df.columns[0]:"D_loss", 
                            df.columns[1]:"D_acc"})
    
    d_losses, d_accs = [], []

    for i, row in df.iterrows():
        d_loss, d_acc = float(row[0].replace(',', '')), float(row[1].replace(']', ''))
        d_losses.append(d_loss); d_accs.append(d_acc)

    new_df = pd.DataFrame({
        "d_loss":d_losses,
        "d_acc":d_accs
    })
    
    return new_df
```
