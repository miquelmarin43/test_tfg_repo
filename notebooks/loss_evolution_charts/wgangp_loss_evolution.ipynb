```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Funciones para convertir txt a csv y procesar columnas

```python
def replace_characters_wgan(path):
    df = pd.read_csv(path, encoding='unicode_escape', sep=" ", header=None)
    df = df.drop(columns=[0, 1, 2, 4, 5])
    df = df.rename(columns={df.columns[0]:"D_loss", df.columns[1]:"G_loss"})
    
    d_losses, g_losses = [], []

    for i, row in df.iterrows():
        d_loss, g_loss = float(row[0].replace(']', '')), float(row[1].replace(']', ''))
        d_losses.append(d_loss); g_losses.append(g_loss)

    new_df = pd.DataFrame({
        "d_loss":d_losses,
        "g_loss":g_losses
    })
    
    return new_df
```
